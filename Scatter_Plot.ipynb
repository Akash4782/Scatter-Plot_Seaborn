```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x = "bill_length_mm" , y = "bill_depth_mm" , data = var)
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm" , y = "bill_depth_mm" , data = var , hue = "sex")
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm" , y = "bill_depth_mm" , data = var , hue = "sex" , style = "sex")
plt.show()
```

```python
var1 = sns.load_dataset("penguins").head(50)
var1
```

```python
sns.scatterplot(x = "bill_length_mm" , y = "bill_depth_mm" , data = var1 , hue = "sex" , style = "sex" , 
               size = "sex")
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm" , y = "bill_depth_mm" , data = var1 , hue = "sex" , style = "sex" , 
               size = "sex" , sizes = (200,40))
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm" , y = "bill_depth_mm" , data = var1 , hue = "sex" , style = "sex" , 
               size = "sex" , sizes = (200,40) , palette = 'gist_ncar')
plt.show()
```

```python
m = {"Male":"*" , "Female":"o}
sns.scatterplot(x = "bill_length_mm" , y = "bill_depth_mm" , data = var1 , hue = "sex" , style = "sex" , 
               size = "sex" , sizes = (200,40) , palette = 'gist_ncar' , m)
plt.show()
```
